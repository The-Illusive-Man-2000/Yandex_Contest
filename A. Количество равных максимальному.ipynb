```python
with open('input.txt', 'r') as fin:
    #x = fin.readline()  # число из первой строки
    lines = fin.readlines()  # все остальные строки

lines = lines[:len(lines)-1]

max = -1
counter = 0
for element in lines:
    element = int(element[:len(element)-1])

    if element > max:
        max = element
        counter = 1
    elif element == max:
        counter +=1

print(counter)
```
