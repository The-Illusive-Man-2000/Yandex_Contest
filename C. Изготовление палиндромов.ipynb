```python
l = list(input())

if len(l) == 1:
    result = 0

if len(l) == 2:
    if l[0] == l[1]:
        result = 0
    else:
        result = 1

else:
    if len(l)%2 !=0:
        border = len(l)//2
        part_one = l[0:border]
        part_two = l[border+1:]
        n = len(part_one)
        counter = 0
        for i in range(n):
            if part_one[i] != part_two[n-1-i]:
                counter +=1
        result = counter

    else:
        border = len(l)//2
        part_one = l[0:border]
        part_two = l[border:]
        #print(part_one)
        #print(part_two)
        n = len(part_one)
        counter = 0
        for i in range(n):
            if part_one[i] != part_two[n-1-i]:
                counter +=1
        result = counter



print(result)
```
